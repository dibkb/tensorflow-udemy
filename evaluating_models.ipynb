```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# prepare dataset
X = tf.range(-100,100,2)
X = tf.cast(X,dtype=tf.float32)
y = X + 9
X_train = X[:80]
X_test = X[80:]
y_train = y[:80]
y_test = y[80:]
```

```python
def plot_predictions(y_pred,y_train_pred):
  # plt.scatter(X_train,y_train, label="Training",c="g")
  plt.scatter(X_test,y_test, label="Actual Test",color='darkorange')
  plt.scatter(X_test,y_pred, label="Predictions Test",c="c")
  plt.scatter(X_train,y_train, label="Actual Train",c = "m")
  plt.scatter(X_train,y_train_pred, label="Predictions Test")
  plt.legend()
```

Model 1

```python
tf.random.set_seed(42)
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model_1.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(),metrics = ["mae"])
model_1.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100,verbose=0)
```

```python
y_pred_1 = model_1.predict(X_test)
y_pred_train_1 = model_1.predict(X_train)
plot_predictions(y_pred_1,y_pred_train_1)
```

Model 2
*(using 1 Dense hidden layer)*

```python
tf.random.set_seed(42)
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(1)
])
model_2.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(),metrics = ["mae"])
model_2.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100,verbose=0)
y_pred_2 = model_2.predict(X_test)
y_pred_2_train = model_2.predict(X_train)
plot_predictions(y_pred_2,y_pred_2_train)
```

Model 3
*(using 2 Dense layers)*

```python
tf.random.set_seed(42)
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1)
])
model_3.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(),metrics = ["mae"])
model_3.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100,verbose=0)
y_pred_3 = model_3.predict(X_test)
y_pred_3_train = model_3.predict(X_train)
plot_predictions(y_pred_3,y_pred_3_train)
```

Model 4

```python
tf.random.set_seed(42)
model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation="relu"),
    tf.keras.layers.Dropout(0.5),  # Adding dropout for regularization
    tf.keras.layers.Dense(32, activation='relu'),  # Reducing the number of neurons in the hidden layer
    #  tf.keras.layers.BatchNormalization(),  # Adding batch normalization for improved training stability
    tf.keras.layers.Dropout(0.3),  # Adding another dropout layer
    tf.keras.layers.Dense(1)
])
model_4.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(learning_rate=0.001),metrics = ["mae"])
model_4.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100,verbose=0)
y_pred_4 = model_4.predict(X_test)
y_pred_4_train = model_4.predict(X_train)
plot_predictions(y_pred_4,y_pred_4_train)
```

```python
tf.random.set_seed(42)
model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation="relu"),
    tf.keras.layers.Dropout(0.5),  # Adding dropout for regularization
    tf.keras.layers.Dense(32, activation='relu'),  # Reducing the number of neurons in the hidden layer
     tf.keras.layers.BatchNormalization(),  # Adding batch normalization for improved training stability
    tf.keras.layers.Dropout(0.3),  # Adding another dropout layer
    tf.keras.layers.Dense(1)
])
model_5.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(learning_rate=0.01),metrics = ["mae"])
model_5.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100,verbose=0)
y_pred_5 = model_5.predict(X_test)
y_pred_5_train = model_5.predict(X_train)
plot_predictions(y_pred_5,y_pred_5_train)
```
